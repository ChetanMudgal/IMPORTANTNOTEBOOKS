# Visualization

```python
sns.pairplot(iris,hue='species',size=2.5,aspect=1)
plt.show()
```

```python
iris.hist(edgecolor='black',linewidth=2,figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species',y='petal_length',hue='species',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='petal_width',hue='species',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='sepal_length',hue='species',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='sepal_width',hue='species',data=iris)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.boxplot(x='species',y='petal_length',hue='species',data=iris)
plt.subplot(2,2,2)
sns.boxplot(x='species',y='petal_width',hue='species',data=iris)
plt.subplot(2,2,3)
sns.boxplot(x='species',y='sepal_length',hue='species',data=iris)
plt.subplot(2,2,4)
sns.boxplot(x='species',y='sepal_width',hue='species',data=iris)
plt.show()
```
